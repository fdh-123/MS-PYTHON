#Logistic Regression

 1)How does logistic regression differ from linear regression in terms of its application?

Logistic regression is machine learning algorithm used for classification ,mainly for binary classification. That is, there will be only two outcomes. Where as linear regression is used for regression tasks that is for predicting a continous value.


2- What is the sigmoid function, and why is it used in logistic regression?

In Logistic regression  we are using linear function and sigmoid function.

Logistic regression is used for binary classification, where the goal is to predict probabilities for two classes (e.g., 0 or 1).The sigmoid function converts the linear output (z) of the model into a probability value between 0 and 1.This makes it possible to interpret the output as the likelihood of belonging to a particular class.

A probability threshold (e.g., 0.5) is applied to classify outputs:
If Sigmoid(𝑧)>0.5,predict class 1.
f Sigmoid(z)≤0.5, predict class 0.

3- What is multicollinearity problem and how you solve the multicollinearity problem ?

Multicollinearity occurs when independent variables in a regression model are highly correlated, making it hard to determine their individual effects and leading to unreliable coefficients.

How to Solve Multicollinearity:

Remove one of the correlated variables.

Combine variables using techniques like PCA.

Use L2 Regularization (Ridge Regression).

Check VIF (Variance Inflation Factor) and drop variables with high VIF.







#Decision Tree

1- How does a decision tree algorithm work, and how does it make predictions?

Decision tree is a machine learning algorithm used for both regression and classification tasks.It is also called Rule based algorithm.We are making predictions based on a tree.

To make a prediction, a new data point is passed through the tree starting from the root node.The algorithm starts with the entire dataset and identifies the best feature to split the data.

The "best feature" is chosen based on a criterion such as Information Gain (for classification) or Reduction in Variance (for regression).

At each node, the algorithm evaluates the condition based on the data point's feature value and follows the corresponding branch.

This process continues until the data point reaches a leaf node.
The class label associated with that leaf node is the prediction.

2- What are the criteria used to split a node in a decision tree (e.g., Gini impurity, Information Gain

Gini Impurity: Measures how often a randomly chosen element from the dataset would be incorrectly classified. It will check how much the feature is corelated with the label.

Gini=1-summation(p_i)^2

Entropy (Information Gain): Measures the reduction in uncertainty or impurity after the split.It will check how much our features are uncorelated with the label.

entropy=− ∑p i⋅log 2(p i)


3-What is overfitting in decision trees, and how can it be prevented?

Overfitting in decision trees occurs when the model becomes too complex and learns not only the patterns in the training data but also the noise and random fluctuations. This leads to excellent performance on the training data but poor generalization to unseen data.

Pruning simplifies the tree by removing unnecessary branches, which reduces the complexity and prevents overfitting.

1. Pruning the Tree
Pruning simplifies the tree by removing unnecessary branches, which reduces the complexity and prevents overfitting.

Pre-pruning (Early Stopping):

Stop growing the tree early by setting criteria to limit its size.
Common pre-pruning criteria:
Maximum Depth: Limit the depth of the tree (e.g., max_depth parameter in scikit-learn).
Minimum Samples per Split: Require a minimum number of samples in a node to split further (e.g., min_samples_split).
Minimum Samples per Leaf: Set a minimum number of samples a leaf node must contain (e.g., min_samples_leaf).


Post-pruning:

Grow the tree fully, then prune it back by removing branches that contribute little to predictive performance.
Post-pruning methods are less common in modern libraries like scikit-learn but are used in algorithms such as CART.

We can also use ensemble techniques and cross validation to improve model performance and reduce overfitting.

4- How do you decide the maximum depth of a decision tree?

Start Small: Begin with a small maximum depth (e.g., 3–5) and gradually increase it if the model performs too poorly.

Cross-Validation: Use cross-validation to test different depths. The depth that gives the best performance on the validation data is often the optimal choice.

Avoid Overfitting: If you set the depth too high, the tree might memorize the training data (overfitting). If you set it too low, the tree may be too simple (underfitting).

Visualize: Plot training and validation accuracy at different depths. Choose a depth where validation accuracy is highest and stable, not increasing anymore.

#Ensemble Techniques

1- What is an ensemble learning technique, and why is it used in machine learning?

Ensemble learning is a machine learning technique that combines multiple models (often called base learners or weak learners) to improve the overall performance and predictive accuracy of the system. Instead of relying on a single model, ensemble methods aggregate the predictions of several models to produce a more robust and accurate result.



2- What is bootstrap=True in Ensemble technique ?

Sampling with Replacement: Each model gets its own random sample of the data, but some data points can be repeated (i.e., sampled multiple times).

3-what is pasting?

Pasting: Unlike bagging, where the data is sampled with replacement (meaning some samples can be repeated), pasting involves sampling without replacement. This means that when training each base model in the ensemble, each model gets a random subset of data without any repetitions.

4-What is OOB method?

The Out-of-Bag (OOB) method is used to estimate the performance of an ensemble model (like a Random Forest) without needing a separate test set.

```python
#using Ridge regression model
```

```python
ridgecv.alpha_
```
