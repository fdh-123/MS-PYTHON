```python
#22)Extract the year, month, and day from the date column and add them as new columns
```

```python
dataz = {'date': ['2024-11-30', '2023-10-15', '2022-09-20']}
dfz= pd.DataFrame(dataz)

```

```python
dfz['date'] = pd.to_datetime(dfz['date'])
dfz['year'] = dfz['date'].dt.year
dfz['month'] = dfz['date'].dt.month
dfz['day'] = dfz['date'].dt.day

# Display the DataFrame
print(dfz)
```

```python
#24)calculate the time difference in days between two columns,start_date and end date
```

```python
datao = { 'start_date': ['2024-01-01', '2024-02-15', '2024-03-10'], 'end_date': ['2024-01-10', '2024-02-20', '2024-03-25']}

dfo = pd.DataFrame(datao)
dfo
```

```python
dfo['start_date'] = pd.to_datetime(dfo['start_date'])
dfo['end_date'] = pd.to_datetime(dfo['end_date'])
dfo['days_difference'] = (dfo['end_date'] - dfo['start_date']).dt.days
print(dfo)
```

```python
#25) Group  data by the year from date column and calculate the total sales for each year
```

```python
data82 = {'date': ['2022-01-15', '2022-05-20', '2023-02-10', '2023-07-25'],'sales': [100, 200, 150, 300]}
dfm = pd.DataFrame(data82)
```

```python
dfm['date'] = pd.to_datetime(dfm['date'])
dfm['year'] = dfm['date'].dt.year
yearly_sales = dfm.groupby('year')['sales'].sum().reset_index()

print(yearly_sales)
```
